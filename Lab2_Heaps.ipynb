```python
%matplotlib inline
import seaborn
import matplotlib.pyplot as plt
import re
```

```python
contents = open('data/LEE, Harper - To Kill A Mockingbird.txt').read()
```

```python
cleaned_dummy = re.sub('[\,\.\;\n\"\?\!]+', ' ', contents.lower().replace("\'", '')).split()
```

```python
def get_dict_size(text, n):
    return len(set(text[:n]))
```

```python
x = range(1, 80000, 1000)
y = [get_dict_size(cleaned_dummy, i) for i in x]
plt.plot(x, y)
```
